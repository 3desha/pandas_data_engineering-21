## Load the JSON file that you created in the previous exercise into a dataframe.

```python
import pandas as pd
df = pd.read_json("../data/auto.json")
df = df.set_index('CarNumber')
```

## Make the following selections:

```python
df[(df['Fines']>2100)]
```

```python
df.groupbycount()
```

```python
df[(df['Fines']>2100) & (df['Refund']==2)]
```

```python
models = ['Focus', 'Corolla']
df[(df['Model'].isin(models))]
```

```python
df.groupby(['Make', 'Model']).agg('Fines').count()
```

```python
numbers = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(numbers)]
```

## Make the aggregations with the make and model.

```python
df.groupby('Model')['Fines'].median()
```

```python
df.groupby(['Make','Model'])['Fines'].median()
```

```python
df.groupby('Make')['Fines'].median()
```

```python
df.groupby(['Make', 'Model']).agg('Fines').count()
```

```python
df.groupby(['Make','Model'])['Fines'].agg('count')
```

```python
df.groupby(['Make','Model'])['Fines'].max()
```

```python
df.groupby(['Make','Model'])['Fines'].min()
```

```python
df.groupby(['Make','Model'])['Fines'].std()
```

```python
df.groupby('CarNumber')['Fines'].count().sort_values(ascending=False)
```

```python
most_count = df.groupby('CarNumber')['Fines'].count().sort_values(ascending=False).index[0]
df[df.index==most_count]
```

```python
df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False)
```

```python
most_sum = df.groupby('CarNumber')['Fines'].sum().sort_values(ascending=False).index[0]
df[df.index==most_sum]
```
