## Read the JSON file that you saved in ex02.

```python
import pandas as pd
import numpy as np
import requests
df = pd.read_json("../data/auto.json")
pd.options.display.float_format = '{:.2f}'.format
df
```

## Enrich the dataframe using a sample from that dataframe.

```python
cars = df[['CarNumber', 'Make', 'Model']].drop_duplicates()
sample_df = cars.sample(n=200, random_state = 21, replace=True).reset_index(drop=True)
sample_df['Refund'] = df['Refund'].sample(n=200, random_state=21, replace=True).reset_index(drop=True).values
sample_df['Fines'] = df['Fines'].sample(n=200,random_state=21, replace=True).reset_index(drop=True).values
concat_rows = pd.concat([df, sample_df], ignore_index=True)
concat_rows
```

```python
concat_rows.count()
```

## Enrich the concat_rows dataframe with a new column containing generated data.

```python
np.random.seed(21)
year = pd.Series(np.random.randint(1980,2020, size = len(concat_rows)))
fines = concat_rows
fines['Year'] = year
fines
```

## Enrich the dataframe with data from another dataframe.

```python
surnames_data = pd.read_json("../../datasets/surname.json")
header = surnames_data.iloc[0] 
data = surnames_data.iloc[1:].reset_index(drop=True)  

surnames_df = pd.DataFrame(data.values, columns=header)
surnames_df
SURNAME = pd.Series(surnames_df['NAME'].sample(n=len(concat_rows), random_state = 21, replace=True).reset_index(drop=True))
owners = pd.concat([concat_rows['CarNumber'],SURNAME], axis=1)
owners.rename(columns={'NAME': 'SURNAME'}, inplace=True)
owners
```

```python
new_rows_df = pd.DataFrame({'CarNumber':['A159ТЕRUS','В123ВАRUS','D292FRRUS', 'R273RERUS','D326RUS'],'Refund':[2,1,2,2,1], 'Fines':[1100,5000,4400,3000,2000],
                            'Make':['Toyota', 'Skoda', 'Ford',' Volkswagen','Ford'], 'Model':['Camry', 'Octavia','Focus', 'Polo','Focus'], 'Year':[2012,2016,2014,2018,2008]})
fines = pd.concat([fines,new_rows_df],ignore_index=True)
fines

```

```python
fines.count()
```

```python
owners = owners.iloc[:-20]
```

```python
new_owners = pd.DataFrame({'CarNumber':['G323XMRUS', 'A121ETRUS', 'H637OPRUS'], 'SURNAME':['JORDAN', 'JAMES', 'SCOTT']})
owners = pd.concat([owners, new_owners], ignore_index = True)
owners
```

```python
all_numbers = pd.merge(owners,fines, on='CarNumber', how='outer')#все номера
all_numbers
```

```python
in_both_numbers = pd.merge(owners,fines, how='inner')
in_both_numbers
```

```python
only_in_fines = fines.merge(owners, how='left')
only_in_fines
```

```python
only_in_owners = fines.merge(owners,how='right')
only_in_owners
```

## Create a pivot table from the fines dataframe. It should look like this (the values are the sums of the fines), but with all the years. The values may be different for you.

```python
pivot_table = fines.pivot_table(
    index=['Make', 'Model'],  # группировка по марке и модели
    columns='Year',           # годы становятся столбцами
    values='Fines',     # значения для агрегации
    aggfunc='sum',          # считаем количество штрафов
    fill_value=0              # заполняем нулями отсутствующие значения
)
pivot_table
```
