## Downcasting:

```python
fines.info(memory_usage='deep')
```

```python
new_fines = fines.copy()
for col in new_fines:
    if new_fines[col].dtype =='float64':
        new_fines[col] = new_fines[col].astype('float32')
    else:
        continue
        
        
new_fines.info(memory_usage='deep')
```

```python
import numpy as np
for col in new_fines.columns:
    if new_fines[col].dtype == 'int64':
        # Определяем минимальный возможный тип
        col_min = new_fines[col].min()
        col_max = new_fines[col].max()
        
        if col_min >= 0:  
            if col_max <= np.iinfo(np.uint8).max:
                new_fines[col] = new_fines[col].astype('uint8')
            elif col_max <= np.iinfo(np.uint16).max:
                new_fines[col] =  new_fines[col].astype('uint16')
            elif col_max <= np.iinfo(np.uint32).max:
                new_fines[col] = new_fines[col].astype('uint32')
        else:  
            if (col_min >= np.iinfo(np.int8).min and 
                col_max <= np.iinfo(np.int8).max):
                new_fines[col] = new_fines[col].astype('int8')
            elif (col_min >= np.iinfo(np.int16).min and 
                  col_max <= np.iinfo(np.int16).max):
                new_fines[col] = new_fines[col].astype('int16')
            elif (col_min >= np.iinfo(np.int32).min and 
                  col_max <= np.iinfo(np.int32).max):
                new_fines[col] = new_fines[col].astype('int32')

new_fines.info(memory_usage='deep')
```
