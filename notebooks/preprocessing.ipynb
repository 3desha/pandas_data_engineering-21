## Download and read the CSV file, making ID the index column.

```python
import pandas as pd
df = pd.read_csv('../data/auto.csv', index_col = "ID")
df
```

```python
df2 = pd.read_json('../data/auto.json', orient='records')
```

## Drop the duplicates, taking into account only the following columns: CarNumber, Make_n_Model, and Fines.

```python
df = df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep = 'last')
df.count()
```

```python
df2['Refund'].mean()
```

## Work with missing values.

```python
df.isnull().sum()
df = df.dropna(axis=1, thresh =len(df)-500)
df['Refund'] = df['Refund'].ffill()
df['Fines'].fillna(df['Fines'].mean(), inplace=True)
df.isnull().sum() 
```

## Split and parse the make and model

```python
df['Make'] = df['Make_n_model'].apply(lambda x: x.split()[0] if pd.notna(x) else '')
df['Model'] = df['Make_n_model'].apply(lambda x: ' '.join(x.split()[1:]) if pd.notna(x) else '')
df = df.drop('Make_n_model', axis=1)
```

```python
df.to_json('auto.json', orient='records', indent=None, force_ascii=False)
```
